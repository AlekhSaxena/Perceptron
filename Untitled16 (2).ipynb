```python
from xturing.datasets import InstructionDataset
from xturing.models import BaseModel
```

```python
import json
from datasets import Dataset, DatasetDict
```

```python
def preprocess_alpaca_json_data(alpaca_dataset_path):
    alcapa_data=json.load(open(alcapa_dataset_file))
    instructions=[]
    inputs=[]
    outputs=[]

    for data in alcapa_data:
        instructions.append(data['instruction'])
        inputs.append(data['input'])
        outputs.append(data['output'])

    data_dict={
        "train":{"instruction":instructions,"text":inputs,
                 "target":outputs}
    }

    dataset=DatasetDict()
    for k,v in data_dict.items():
        dataset[k]=Dataset.from_dict(v)
    dataset.save_to_disk(str("alcapa_dataset_path"))
```

```python
preprocess_alpaca_json_data(alpaca_dataset_path='/content/drive/MyDrive/alpaca_plus.json')
```
